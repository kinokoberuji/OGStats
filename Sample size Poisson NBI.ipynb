## Giới thiệu

Hiện nay, hỗ trợ sinh sản là một chuyên đề khoa học đang nhận được nhiều sự quan tâm trong giới Sản Phụ khoa. Có thể nhận ra một điểm thú vị trong những nghiên cứu lâm sàng về kích thích buồng trứng, khởi phát trưởng thành noãn và thụ tinh nhân tạo, đó là nhiều kết cục cần khảo sát có bản chất là số đếm, thí dụ số noãn chọc hút được, số noãn trưởng thành, số noãn thụ tinh, số phôi N3/N5, … 

Những kết cục này thường chỉ được phân tích bằng phương pháp thống kê đơn biến sơ cấp, thí dụ kiểm định so sánh trung bình/trung vị, thậm chí một vài tác giả áp dụng sai kiểm định t-student dù đây là những biến số rời rạc và chắc chắn không thể có phân phối chuẩn.

Phương pháp suy luận thống kê chính xác và phù hợp nhất cho kết cục có bản chất số đếm là một phân tích hồi quy theo phân phối Poisson, hoặc tốt hơn là Negative Binomial (nhị thức âm) nếu dữ liệu quá phân tán. Đây là những quy luật phân phối thích hợp cho số đếm. Việc sử dụng mô hình hồi quy còn cho phép hiệu chỉnh cho các hiệp biến như tuổi, BMI, thời gian/nguyên nhân hiếm muộn, mức dự trữ buồng trứng, giá trị hormone AMH,FSH,E2 cơ bản… 

Trong bài này, Nhi xin giới thiệu quy trình tính cỡ mẫu cho bài toán so sánh phân phối Poisson hoặc Negative Binomial giữa 2 phân nhóm can thiệp và đối chứng.

## Bài toán so sánh 2 phân phối Negative Binomial 

Trong bài toán này, ta cần so sánh giá trị trung bình của một kết cục Y có bản chất là số đếm (thí dụ: số lượng noãn trưởng thành) giữa nhóm đối chứng N0 và nhóm can thiệp N1, với giả định Y là một biến ngẫu nhiên theo quy luật phân phối Negative Binomial, được xác định bằng 2 tham số: trung bình $\mu$ và độ phân tán $k$

Cỡ mẫu được ước lượng nhằm tối ưu độ chính xác của một mô hình hồi quy nhị thức âm (negative binomial) nhằm so sánh số lượng noãn trưởng thành giữa 2 phân nhóm, dựa theo phương pháp của Cundill và Alexander (2015). 

Với 2 trường hợp: (1) mô hình sử dụng một hàm liên kết (link function) logarit (thang đo log(Y)) :

$$\sqrt{N}=\frac{Z_{1-\frac{\alpha }{2}}\sqrt{\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)\left(\frac{1}{Q_1}+\frac{1}{Q_0}\right)}+{Z}_{1-\beta}\sqrt{\frac{1}{Q_1}\left(\frac{1}{\mu_1}+\frac{1}{k_1}\right)+\frac{1}{Q_0}\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)}}{ \log \left({\mu}_0\right)- \log \left({\mu}_1\right)}$$

Và (2): khảo sát thang đo tự nhiên của Y (một sự xấp xỉ phân phối chuẩn)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{\frac{1}{Q_1}\left({\mu}_1+\frac{\mu_1^2}{k_1}\right)+\frac{1}{Q_0}\left({\mu}_0+\frac{\mu_0^2}{k_0}\right)}}{\mu_0-{\mu}_1}$$

## Bài toán so sánh 2 phân phối Poisson 

Bài toán nêu trên cũng có thể tiếp cận bằng quy luật phân phối Poisson, và cỡ mẫu cũng có thể ước tính bởi 2 công thức: (1) tương ứng với thang đo log(Y) (hàm liên kết logarit)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{\frac{1}{Q_1}\left(\frac{1}{\mu_1}+\frac{1}{k_1}\right)+\frac{1}{Q_0}\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)}}{ \log \left({\mu}_0\right)- \log \left({\mu}_1\right)}$$

và (2) cho thang đo tự nhiên (xấp xỉ phân phối chuẩn)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{2\left({\mu}_1+{\mu}_0\right)}}{\mu_0-{\mu}_1}$$

Trong đó :

$N$ là tổng cỡ mẫu của cả 2 phân nhóm (Chú ý : đơn vị quan sát là chu kì COS hay IVF, không phải bệnh nhân)

$μ_0$ và $μ_1$ tương ứng với giá trị trung bình của kết cục Y trong nhóm đối chứng (0 : phác đồ tham chiếu) và nhóm can thiệp (1 : phác đồ mới cần khảo sát).  

$Q_0$ và $Q_1$ là tỉ lệ cỡ mẫu của 2 phân nhóm trên tổng cỡ mẫu, cho phép thiết kế thí nghiệm với 2 nhóm không đối xứng về cỡ mẫu. Trong trường hợp 2 phân nhóm đều nhau, $Q_0 = Q_1  = 0.5$

$k_0$ và $k_1$ tương ứng với hệ số phân tán của phân phối nhị thức âm (negative binomial). 

Theo lý thuyết, $k$ là số lần thử nghiệm thất bại, đối nghịch với $mu$ là số lần thử nghiệm thành công. Ta có thể xác định $mu$ và $k$ dễ dàng khi đọc kết quả từ y văn. Thí dụ : nếu kết cục Y là số noãn trưởng thành (MII), thì $mu$ chính là giá trị trung bình được báo cáo cho số lượng noãn MII, còn $k$ = trung bình của số noãn chọc hút được trừ cho $mu$ ($k$ = số noãn không trưởng thành). Tương tự, nếu kết cục là số noãn thụ tinh, thì k = số noãn không thụ tinh

$Z_p$ là bách phân vị thứ p của phân phối chuẩn hóa (standard normal)

$α$ là ngưỡng ý nghĩa thống kê, hay xác suất phạm sai lầm loại 1 (phủ nhận sai một giả thuyết vô hiệu khi mà nó đúng). Theo quy ước, ta chấp nhận $𝛼  = 0,05$

$𝛽$ là xác suất phạm sai lầm loại II (âm tính giả).

$Power = 1 - 𝛽$ ; Lực thống kê, là xác suất phủ nhận chính xác giả thuyết vô hiệu nếu quả thực giả thuyết này sai. Nói cách khác là khả năng phát hiện ra một hiệu ứng khác biệt nếu nó thực sự tồn tại. Theo quy ước, ta thường chấp nhận $power$ = 0,8 (tương ứng $𝛽$ = 0,2).

## Mã lập trình

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm

from typing import Tuple

%config InlineBackend.figure_format = 'png'
```

### Hàm ước tính cỡ mẫu

Ghi chú : dist = ‘POI’ cho phân phối Poisson, ‘NBI’ cho phân phối Negative Binomial,

method = ‘norm’ nếu dùng thang đo tự nhiên (xấp xỉ phân phối chuẩn), hoặc ‘org’ nếu dùng thang đo logarit log(Y)

Quy ước : $\mu_0 > \mu_1$

```python
def sample_size_countdata(mu0: float = None,
                          k0: float = None,
                          mu1: float = None,
                          k1: float = None,
                          q0: float = 0.5,
                          q1: float = 0.5,
                          alpha: float = 0.05,
                          beta: float = 0.2,
                          dist: str = 'NBI',
                          method = 'norm',
                          ):
    
    z1 = norm.ppf(1-alpha/2)
    z2 = norm.ppf(1-beta)
        
    if dist == 'NBI':
        
        if method == 'org':
            
            g1 = z1*np.sqrt(((1/mu0)+(1/k0))*((1/q0)+(1/q1)))
            g2 = z2*np.sqrt((1/q1)*((1/mu1) + (1/k1)) + (1/q0)*((1/mu0)+(1/k0)))
            d = np.log(mu0) - np.log(mu1)

            N = ((g1 + g2)/d)**2

            n0 = N*q0
            n1 = N*q1

            return n0, n1, N
        
        elif method == 'norm':
            
            g1 = (mu1 + ((mu1**2)/k1))/q1
            g2 = (mu0 + ((mu0**2)/k0))/q0
            
            d = mu0 - mu1
            
            N = ((z1 + z2)*(np.sqrt(g1 + g2)))/d
            N = N**2
            
            n0 = N*q0
            n1 = N*q1
            
            return n0, n1, N
        else:
            print("Invalid method")
            return None, None, None
    
    elif dist == 'POI':
        
        g1 = ((1/mu1)+(1/k1))/q1
        g2 = ((1/mu0)+(1/k0))/q0
        
        d = np.log(mu0) - np.log(mu1)
        
        N = ((z1 + z2)*(np.sqrt(g1 + g2)))/d
        N = N**2

        n0 = N*q0
        n1 = N*q1
            
        return n0, n1, N
    
    else:
        print("Invalid distribution")
        return None, None, None
```

### Hàm mô phỏng sự dao động của $\mu$ và $k$

```python
def plot_simulat(mu0: float = None,
                k0: float = None,
                mu1: float = None,
                k1: float = None,
                q0: float = 0.5,
                q1: float = 0.5,
                min_mu0: float = None,
                max_mu0: float = None,
                min_k0: float = None,
                max_k0: float = None,
                min_mu1: float = None,
                max_mu1: float = None,
                min_k1: float = None,
                max_k1: float = None,
                alpha: float = 0.05,
                beta: float = 0.2,
                dist: str = 'NBI',
                method = 'norm'):
    
    sns.set_palette('viridis', 50)
    
    s_mu0 = np.linspace(min_mu0, max_mu0, num = 50)
    s_mu1 = np.linspace(min_mu1, max_mu1, num = 50)
    s_k0 = np.linspace(min_k0, max_k0, num = 50)
    s_k1 = np.linspace(min_k1, max_k1, num = 50)
    
    plt.rc('font', size=15) 
    fig, axs = plt.subplots(2,2, figsize=(14,14))
    
    for m0_i in s_mu0:
        n1_m = [sample_size_countdata(mu0 = m0_i,
                                    k0 = k0_i,
                                    mu1 = mu1,
                                    k1 = k1,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for k0_i in s_k0]
        
        axs[1,0].plot(s_k0, n1_m)
        
    axs[1,0].set_ylabel('Cỡ mẫu N1')
    axs[1,0].set_xlabel('k_0')
    
    for k0_i in s_k0:
        n1_k = [sample_size_countdata(mu0 = m0_i,
                                    k0 = k0_i,
                                    mu1 = mu1,
                                    k1 = k1,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for m0_i in s_mu0]
        
        axs[0,0].plot(s_mu0, n1_k)
    
    axs[0,0].set_ylabel('Cỡ mẫu N1')
    axs[0,0].set_xlabel('mu_0')
        
    for m1_i in s_mu1:
        n2_m = [sample_size_countdata(mu0 = mu0, k0 = k0,
                                    mu1 = m1_i, k1 = k1_i,
                                    q0 = q0, q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for k1_i in s_k1]
        
        axs[1,1].plot(s_k1, n2_m)
        
    axs[1,1].set_ylabel('Cỡ mẫu N2')
    axs[1,1].set_xlabel('k_1')
        
    for k1_i in s_k1:
        n2_k = [sample_size_countdata(mu0 = mu0,
                                    k0 = k0,
                                    mu1 = m1_i,
                                    k1 = k1_i,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for m1_i in s_mu1]
        
        axs[0,1].plot(s_mu1, n2_k)
    
    axs[0,1].set_ylabel('Cỡ mẫu N2')
    axs[0,1].set_xlabel('mu_1')
        
    plt.tight_layout()
    plt.show()
```
