## Gi·ªõi thi·ªáu

Hi·ªán nay, h·ªó tr·ª£ sinh s·∫£n l√† m·ªôt chuy√™n ƒë·ªÅ khoa h·ªçc ƒëang nh·∫≠n ƒë∆∞·ª£c nhi·ªÅu s·ª± quan t√¢m trong gi·ªõi S·∫£n Ph·ª• khoa. C√≥ th·ªÉ nh·∫≠n ra m·ªôt ƒëi·ªÉm th√∫ v·ªã trong nh·ªØng nghi√™n c·ª©u l√¢m s√†ng v·ªÅ k√≠ch th√≠ch bu·ªìng tr·ª©ng, kh·ªüi ph√°t tr∆∞·ªüng th√†nh no√£n v√† th·ª• tinh nh√¢n t·∫°o, ƒë√≥ l√† nhi·ªÅu k·∫øt c·ª•c c·∫ßn kh·∫£o s√°t c√≥ b·∫£n ch·∫•t l√† s·ªë ƒë·∫øm, th√≠ d·ª• s·ªë no√£n ch·ªçc h√∫t ƒë∆∞·ª£c, s·ªë no√£n tr∆∞·ªüng th√†nh, s·ªë no√£n th·ª• tinh, s·ªë ph√¥i N3/N5, ‚Ä¶ 

Nh·ªØng k·∫øt c·ª•c n√†y th∆∞·ªùng ch·ªâ ƒë∆∞·ª£c ph√¢n t√≠ch b·∫±ng ph∆∞∆°ng ph√°p th·ªëng k√™ ƒë∆°n bi·∫øn s∆° c·∫•p, th√≠ d·ª• ki·ªÉm ƒë·ªãnh so s√°nh trung b√¨nh/trung v·ªã, th·∫≠m ch√≠ m·ªôt v√†i t√°c gi·∫£ √°p d·ª•ng sai ki·ªÉm ƒë·ªãnh t-student d√π ƒë√¢y l√† nh·ªØng bi·∫øn s·ªë r·ªùi r·∫°c v√† ch·∫Øc ch·∫Øn kh√¥ng th·ªÉ c√≥ ph√¢n ph·ªëi chu·∫©n.

Ph∆∞∆°ng ph√°p suy lu·∫≠n th·ªëng k√™ ch√≠nh x√°c v√† ph√π h·ª£p nh·∫•t cho k·∫øt c·ª•c c√≥ b·∫£n ch·∫•t s·ªë ƒë·∫øm l√† m·ªôt ph√¢n t√≠ch h·ªìi quy theo ph√¢n ph·ªëi Poisson, ho·∫∑c t·ªët h∆°n l√† Negative Binomial (nh·ªã th·ª©c √¢m) n·∫øu d·ªØ li·ªáu qu√° ph√¢n t√°n. ƒê√¢y l√† nh·ªØng quy lu·∫≠t ph√¢n ph·ªëi th√≠ch h·ª£p cho s·ªë ƒë·∫øm. Vi·ªác s·ª≠ d·ª•ng m√¥ h√¨nh h·ªìi quy c√≤n cho ph√©p hi·ªáu ch·ªânh cho c√°c hi·ªáp bi·∫øn nh∆∞ tu·ªïi, BMI, th·ªùi gian/nguy√™n nh√¢n hi·∫øm mu·ªôn, m·ª©c d·ª± tr·ªØ bu·ªìng tr·ª©ng, gi√° tr·ªã hormone AMH,FSH,E2 c∆° b·∫£n‚Ä¶ 

Trong b√†i n√†y, Nhi xin gi·ªõi thi·ªáu quy tr√¨nh t√≠nh c·ª° m·∫´u cho b√†i to√°n so s√°nh ph√¢n ph·ªëi Poisson ho·∫∑c Negative Binomial gi·ªØa 2 ph√¢n nh√≥m can thi·ªáp v√† ƒë·ªëi ch·ª©ng.

## B√†i to√°n so s√°nh 2 ph√¢n ph·ªëi Negative Binomial 

Trong b√†i to√°n n√†y, ta c·∫ßn so s√°nh gi√° tr·ªã trung b√¨nh c·ªßa m·ªôt k·∫øt c·ª•c Y c√≥ b·∫£n ch·∫•t l√† s·ªë ƒë·∫øm (th√≠ d·ª•: s·ªë l∆∞·ª£ng no√£n tr∆∞·ªüng th√†nh) gi·ªØa nh√≥m ƒë·ªëi ch·ª©ng N0 v√† nh√≥m can thi·ªáp N1, v·ªõi gi·∫£ ƒë·ªãnh Y l√† m·ªôt bi·∫øn ng·∫´u nhi√™n theo quy lu·∫≠t ph√¢n ph·ªëi Negative Binomial, ƒë∆∞·ª£c x√°c ƒë·ªãnh b·∫±ng 2 tham s·ªë: trung b√¨nh $\mu$ v√† ƒë·ªô ph√¢n t√°n $k$

C·ª° m·∫´u ƒë∆∞·ª£c ∆∞·ªõc l∆∞·ª£ng nh·∫±m t·ªëi ∆∞u ƒë·ªô ch√≠nh x√°c c·ªßa m·ªôt m√¥ h√¨nh h·ªìi quy nh·ªã th·ª©c √¢m (negative binomial) nh·∫±m so s√°nh s·ªë l∆∞·ª£ng no√£n tr∆∞·ªüng th√†nh gi·ªØa 2 ph√¢n nh√≥m, d·ª±a theo ph∆∞∆°ng ph√°p c·ªßa Cundill v√† Alexander (2015). 

V·ªõi 2 tr∆∞·ªùng h·ª£p: (1) m√¥ h√¨nh s·ª≠ d·ª•ng m·ªôt h√†m li√™n k·∫øt (link function) logarit (thang ƒëo log(Y)) :

$$\sqrt{N}=\frac{Z_{1-\frac{\alpha }{2}}\sqrt{\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)\left(\frac{1}{Q_1}+\frac{1}{Q_0}\right)}+{Z}_{1-\beta}\sqrt{\frac{1}{Q_1}\left(\frac{1}{\mu_1}+\frac{1}{k_1}\right)+\frac{1}{Q_0}\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)}}{ \log \left({\mu}_0\right)- \log \left({\mu}_1\right)}$$

V√† (2): kh·∫£o s√°t thang ƒëo t·ª± nhi√™n c·ªßa Y (m·ªôt s·ª± x·∫•p x·ªâ ph√¢n ph·ªëi chu·∫©n)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{\frac{1}{Q_1}\left({\mu}_1+\frac{\mu_1^2}{k_1}\right)+\frac{1}{Q_0}\left({\mu}_0+\frac{\mu_0^2}{k_0}\right)}}{\mu_0-{\mu}_1}$$

## B√†i to√°n so s√°nh 2 ph√¢n ph·ªëi Poisson 

B√†i to√°n n√™u tr√™n c≈©ng c√≥ th·ªÉ ti·∫øp c·∫≠n b·∫±ng quy lu·∫≠t ph√¢n ph·ªëi Poisson, v√† c·ª° m·∫´u c≈©ng c√≥ th·ªÉ ∆∞·ªõc t√≠nh b·ªüi 2 c√¥ng th·ª©c: (1) t∆∞∆°ng ·ª©ng v·ªõi thang ƒëo log(Y) (h√†m li√™n k·∫øt logarit)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{\frac{1}{Q_1}\left(\frac{1}{\mu_1}+\frac{1}{k_1}\right)+\frac{1}{Q_0}\left(\frac{1}{\mu_0}+\frac{1}{k_0}\right)}}{ \log \left({\mu}_0\right)- \log \left({\mu}_1\right)}$$

v√† (2) cho thang ƒëo t·ª± nhi√™n (x·∫•p x·ªâ ph√¢n ph·ªëi chu·∫©n)

$$\sqrt{N}=\frac{\left({Z}_{1-\frac{\alpha }{2}}+{Z}_{1-\beta}\right)\sqrt{2\left({\mu}_1+{\mu}_0\right)}}{\mu_0-{\mu}_1}$$

Trong ƒë√≥ :

$N$ l√† t·ªïng c·ª° m·∫´u c·ªßa c·∫£ 2 ph√¢n nh√≥m (Ch√∫ √Ω : ƒë∆°n v·ªã quan s√°t l√† chu k√¨ COS hay IVF, kh√¥ng ph·∫£i b·ªánh nh√¢n)

$Œº_0$ v√† $Œº_1$ t∆∞∆°ng ·ª©ng v·ªõi gi√° tr·ªã trung b√¨nh c·ªßa k·∫øt c·ª•c Y trong nh√≥m ƒë·ªëi ch·ª©ng (0 : ph√°c ƒë·ªì tham chi·∫øu) v√† nh√≥m can thi·ªáp (1 : ph√°c ƒë·ªì m·ªõi c·∫ßn kh·∫£o s√°t).  

$Q_0$ v√† $Q_1$ l√† t·ªâ l·ªá c·ª° m·∫´u c·ªßa 2 ph√¢n nh√≥m tr√™n t·ªïng c·ª° m·∫´u, cho ph√©p thi·∫øt k·∫ø th√≠ nghi·ªám v·ªõi 2 nh√≥m kh√¥ng ƒë·ªëi x·ª©ng v·ªÅ c·ª° m·∫´u. Trong tr∆∞·ªùng h·ª£p 2 ph√¢n nh√≥m ƒë·ªÅu nhau, $Q_0 = Q_1  = 0.5$

$k_0$ v√† $k_1$ t∆∞∆°ng ·ª©ng v·ªõi h·ªá s·ªë ph√¢n t√°n c·ªßa ph√¢n ph·ªëi nh·ªã th·ª©c √¢m (negative binomial). 

Theo l√Ω thuy·∫øt, $k$ l√† s·ªë l·∫ßn th·ª≠ nghi·ªám th·∫•t b·∫°i, ƒë·ªëi ngh·ªãch v·ªõi $mu$ l√† s·ªë l·∫ßn th·ª≠ nghi·ªám th√†nh c√¥ng. Ta c√≥ th·ªÉ x√°c ƒë·ªãnh $mu$ v√† $k$ d·ªÖ d√†ng khi ƒë·ªçc k·∫øt qu·∫£ t·ª´ y vƒÉn. Th√≠ d·ª• : n·∫øu k·∫øt c·ª•c Y l√† s·ªë no√£n tr∆∞·ªüng th√†nh (MII), th√¨ $mu$ ch√≠nh l√† gi√° tr·ªã trung b√¨nh ƒë∆∞·ª£c b√°o c√°o cho s·ªë l∆∞·ª£ng no√£n MII, c√≤n $k$ = trung b√¨nh c·ªßa s·ªë no√£n ch·ªçc h√∫t ƒë∆∞·ª£c tr·ª´ cho $mu$ ($k$ = s·ªë no√£n kh√¥ng tr∆∞·ªüng th√†nh). T∆∞∆°ng t·ª±, n·∫øu k·∫øt c·ª•c l√† s·ªë no√£n th·ª• tinh, th√¨ k = s·ªë no√£n kh√¥ng th·ª• tinh

$Z_p$ l√† b√°ch ph√¢n v·ªã th·ª© p c·ªßa ph√¢n ph·ªëi chu·∫©n h√≥a (standard normal)

$Œ±$ l√† ng∆∞·ª°ng √Ω nghƒ©a th·ªëng k√™, hay x√°c su·∫•t ph·∫°m sai l·∫ßm lo·∫°i 1 (ph·ªß nh·∫≠n sai m·ªôt gi·∫£ thuy·∫øt v√¥ hi·ªáu khi m√† n√≥ ƒë√∫ng). Theo quy ∆∞·ªõc, ta ch·∫•p nh·∫≠n $ùõº  = 0,05$

$ùõΩ$ l√† x√°c su·∫•t ph·∫°m sai l·∫ßm lo·∫°i II (√¢m t√≠nh gi·∫£).

$Power = 1 - ùõΩ$ ; L·ª±c th·ªëng k√™, l√† x√°c su·∫•t ph·ªß nh·∫≠n ch√≠nh x√°c gi·∫£ thuy·∫øt v√¥ hi·ªáu n·∫øu qu·∫£ th·ª±c gi·∫£ thuy·∫øt n√†y sai. N√≥i c√°ch kh√°c l√† kh·∫£ nƒÉng ph√°t hi·ªán ra m·ªôt hi·ªáu ·ª©ng kh√°c bi·ªát n·∫øu n√≥ th·ª±c s·ª± t·ªìn t·∫°i. Theo quy ∆∞·ªõc, ta th∆∞·ªùng ch·∫•p nh·∫≠n $power$ = 0,8 (t∆∞∆°ng ·ª©ng $ùõΩ$ = 0,2).

## M√£ l·∫≠p tr√¨nh

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
from scipy.stats import norm

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm

from typing import Tuple

%config InlineBackend.figure_format = 'png'
```

### H√†m ∆∞·ªõc t√≠nh c·ª° m·∫´u

Ghi ch√∫ : dist = ‚ÄòPOI‚Äô cho ph√¢n ph·ªëi Poisson, ‚ÄòNBI‚Äô cho ph√¢n ph·ªëi Negative Binomial,

method = ‚Äònorm‚Äô n·∫øu d√πng thang ƒëo t·ª± nhi√™n (x·∫•p x·ªâ ph√¢n ph·ªëi chu·∫©n), ho·∫∑c ‚Äòorg‚Äô n·∫øu d√πng thang ƒëo logarit log(Y)

Quy ∆∞·ªõc : $\mu_0 > \mu_1$

```python
def sample_size_countdata(mu0: float = None,
                          k0: float = None,
                          mu1: float = None,
                          k1: float = None,
                          q0: float = 0.5,
                          q1: float = 0.5,
                          alpha: float = 0.05,
                          beta: float = 0.2,
                          dist: str = 'NBI',
                          method = 'norm',
                          ):
    
    z1 = norm.ppf(1-alpha/2)
    z2 = norm.ppf(1-beta)
        
    if dist == 'NBI':
        
        if method == 'org':
            
            g1 = z1*np.sqrt(((1/mu0)+(1/k0))*((1/q0)+(1/q1)))
            g2 = z2*np.sqrt((1/q1)*((1/mu1) + (1/k1)) + (1/q0)*((1/mu0)+(1/k0)))
            d = np.log(mu0) - np.log(mu1)

            N = ((g1 + g2)/d)**2

            n0 = N*q0
            n1 = N*q1

            return n0, n1, N
        
        elif method == 'norm':
            
            g1 = (mu1 + ((mu1**2)/k1))/q1
            g2 = (mu0 + ((mu0**2)/k0))/q0
            
            d = mu0 - mu1
            
            N = ((z1 + z2)*(np.sqrt(g1 + g2)))/d
            N = N**2
            
            n0 = N*q0
            n1 = N*q1
            
            return n0, n1, N
        else:
            print("Invalid method")
            return None, None, None
    
    elif dist == 'POI':
        
        g1 = ((1/mu1)+(1/k1))/q1
        g2 = ((1/mu0)+(1/k0))/q0
        
        d = np.log(mu0) - np.log(mu1)
        
        N = ((z1 + z2)*(np.sqrt(g1 + g2)))/d
        N = N**2

        n0 = N*q0
        n1 = N*q1
            
        return n0, n1, N
    
    else:
        print("Invalid distribution")
        return None, None, None
```

### H√†m m√¥ ph·ªèng s·ª± dao ƒë·ªông c·ªßa $\mu$ v√† $k$

```python
def plot_simulat(mu0: float = None,
                k0: float = None,
                mu1: float = None,
                k1: float = None,
                q0: float = 0.5,
                q1: float = 0.5,
                min_mu0: float = None,
                max_mu0: float = None,
                min_k0: float = None,
                max_k0: float = None,
                min_mu1: float = None,
                max_mu1: float = None,
                min_k1: float = None,
                max_k1: float = None,
                alpha: float = 0.05,
                beta: float = 0.2,
                dist: str = 'NBI',
                method = 'norm'):
    
    sns.set_palette('viridis', 50)
    
    s_mu0 = np.linspace(min_mu0, max_mu0, num = 50)
    s_mu1 = np.linspace(min_mu1, max_mu1, num = 50)
    s_k0 = np.linspace(min_k0, max_k0, num = 50)
    s_k1 = np.linspace(min_k1, max_k1, num = 50)
    
    plt.rc('font', size=15) 
    fig, axs = plt.subplots(2,2, figsize=(14,14))
    
    for m0_i in s_mu0:
        n1_m = [sample_size_countdata(mu0 = m0_i,
                                    k0 = k0_i,
                                    mu1 = mu1,
                                    k1 = k1,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for k0_i in s_k0]
        
        axs[1,0].plot(s_k0, n1_m)
        
    axs[1,0].set_ylabel('C·ª° m·∫´u N1')
    axs[1,0].set_xlabel('k_0')
    
    for k0_i in s_k0:
        n1_k = [sample_size_countdata(mu0 = m0_i,
                                    k0 = k0_i,
                                    mu1 = mu1,
                                    k1 = k1,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for m0_i in s_mu0]
        
        axs[0,0].plot(s_mu0, n1_k)
    
    axs[0,0].set_ylabel('C·ª° m·∫´u N1')
    axs[0,0].set_xlabel('mu_0')
        
    for m1_i in s_mu1:
        n2_m = [sample_size_countdata(mu0 = mu0, k0 = k0,
                                    mu1 = m1_i, k1 = k1_i,
                                    q0 = q0, q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for k1_i in s_k1]
        
        axs[1,1].plot(s_k1, n2_m)
        
    axs[1,1].set_ylabel('C·ª° m·∫´u N2')
    axs[1,1].set_xlabel('k_1')
        
    for k1_i in s_k1:
        n2_k = [sample_size_countdata(mu0 = mu0,
                                    k0 = k0,
                                    mu1 = m1_i,
                                    k1 = k1_i,
                                    q0 = q0,
                                    q1 = q1,
                                    alpha = alpha,
                                    beta = beta,
                                    dist = dist,
                                    method = method)[0] for m1_i in s_mu1]
        
        axs[0,1].plot(s_mu1, n2_k)
    
    axs[0,1].set_ylabel('C·ª° m·∫´u N2')
    axs[0,1].set_xlabel('mu_1')
        
    plt.tight_layout()
    plt.show()
```
